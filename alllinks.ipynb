```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
```

```python
def getSoup(link):
    r = requests.get(link)
    r.encoding = 'UTF-8'
    return BeautifulSoup(r.text,'lxml')
```

```python
def getAllLinks(link):
    soup = getSoup(link)
    tds = soup.findAll('td', {'align':'center'})
    return [td.find('a')['href'] for td in tds]

links = getAllLinks('https://www.f1-fansite.com/f1-results/')
```

```python
def getName(link):
    soup = getSoup(link)
    return soup.find('h2').text
[getName(link) for link in links]
```

```python
def GetTable(link):
    soup = getSoup(link)
    return soup.find("table", {"class":"motor-sport-results msr_season_driver_results"})
```
