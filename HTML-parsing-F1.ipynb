```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
```

```python
os.chdir("Desktop/")
```

```python
url = "https://www.f1-fansite.com/f1-results/1950-f1-championship-standings/"
r = requests.get(url)
text = r.text
soup = BeautifulSoup(text)
```

```python
print(soup.prettify())
```

```python
My_table = soup.find("table", {"class":"motor-sport-results msr_season_summary tablesorter"})
```

```python
table_body = My_table.find('tbody')
table_body
```

```python
data = []
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for  ele in cols]
    data.append([ele for ele in cols if ele])
```

```python
data
```

```python
tab = pd.DataFrame(data)
tab
```

```python
table_head = My_table.find('thead')
table_head
```

```python
headers = []
rows = table_head.find_all('tr')
rows
```
